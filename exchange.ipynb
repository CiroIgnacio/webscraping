## Conversor de monedas

```python
base_url = 'https://api.exchangeratesapi.io'
base_url
```

```python
fecha = input("Ingrese la fecha de cambio en formato 'AAAA-DD-MM', para hacerlo con el cambio actual ingrese 'hoy': ")
base = input("Convertir desde: ")
curr = input("Hacia: ")
cant = float(input(f"Cuántos {base} desea convertir: "))

if fecha.lower() == 'hoy':
    fecha_latest = 'latest'
else:
    fecha_latest = fecha
url = base_url + "/" + fecha_latest + "?base=" + base + "&symbols=" + curr
response = requests.get(url)


if response.ok is False:
    print(f"\n Error: {response.status_code}")
    print(response.json()['error'])

else:
    data = response.json()
    print(data)
    proporcion = data['rates'][curr]
    
    resultado = cant*proporcion

    print(f"{cant} {base} son iguales a {resultado} {curr} en la fecha {fecha}")
    

```
