### ITUNES API 

```python
base_url = "https://itunes.apple.com/search"
```

```python
import requests
```

PASAR PAR√ÅMETROS USANDO GET METHOD


```python
r = requests.get(base_url, params = {"term": "nirvana", "country": "us", "limit": 200})
r.url
```

```python
import json
info = r.json()
print(json.dumps(info, indent = 4))
```
